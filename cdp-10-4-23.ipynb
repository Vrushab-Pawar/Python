```python
N = [1,2,45,77,65,75,35,24,2,19,8,15,22,27,13,18]
Max = []
Min = []

while N:
    max_val = max(N)
    Max.append(max_val)
    N.remove(max_val)
    
    
    min_val = min(N)
    Min.append(min_val)
    N.remove(min_val)
    

        
print(Max,'\n',Min)
```
